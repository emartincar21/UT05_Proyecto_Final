```python
import pandas as pd
import re

# 1. Cargar el Mega-CSV en bruto
print("Cargando datos...")
df_bruto = pd.read_csv('COPIA_SPANISH_HOUSES.csv', low_memory=False)
```

```python
# 2. Seleccionar las columnas estrella (SIN renta_media_prov)
columnas_estrella = [
    'loc_zone', # Localización
    'm2_real', 'room_num', 'bath_num', 'condition', 'floor', # Físicas
    'lift', 'garage', 'swimming_pool', 'air_conditioner', 'terrace', # Extras
    'price' # Objetivo
]
df_red = df_bruto[columnas_estrella].copy()

# 3. LIMPIEZA DE NÚMEROS Y NULOS BÁSICA
# Ya no incluimos renta_media_prov
for col in ['price', 'm2_real', 'room_num', 'bath_num']:
    df_red[col] = pd.to_numeric(df_red[col], errors='coerce')

df_red = df_red.dropna(subset=['price', 'm2_real', 'loc_zone'])
```

```python
df_red.head()
```

```python

# 4. TRATAMIENTO DE LOCALIZACIÓN (EL TRUCO MAGISTRAL)
# Calculamos el precio medio de cada zona y lo guardamos en un diccionario
precio_medio_por_zona = df_red.groupby('loc_zone')['price'].mean().to_dict()

# Reemplazamos el nombre de la zona (texto) por su precio medio (número)
# Lo llamamos 'loc_zone_value'
df_red['loc_zone_value'] = df_red['loc_zone'].map(precio_medio_por_zona)


```

```python

# 5. CONVERSIÓN DE TEXTO A NÚMERO (Floor y Condition)
# Extraer número de planta
def limpiar_planta(x):
    if pd.isna(x): return 0
    nums = re.findall(r'\d+', str(x))
    return int(nums[0]) if nums else 0
df_red['floor'] = df_red['floor'].apply(limpiar_planta)

# Mapear estado
mapping_cond = {'segunda mano/para reformar': 0, 'segunda mano/buen estado': 1, 'promoción de obra nueva': 2}
df_red['condition'] = df_red['condition'].map(mapping_cond).fillna(1).astype(int)

# --- NUEVA VERSIÓN PARA LOS EXTRAS ---
extras = ['lift', 'garage', 'swimming_pool', 'air_conditioner', 'terrace']

def limpiar_extras_letras(x):
    # 1. Si está totalmente vacío (NaN), ponemos un 0
    if pd.isna(x): 
        return 0
    
    # 2. Convertimos a texto para analizarlo
    texto = str(x).strip()
    
    # 3. Si el dato era explícitamente un cero ("0" o "0.0"), lo mantenemos como 0
    if texto in ['0', '0.0']:
        return 0
        
    # 4. Si hay CUALQUIER otra cosa (letras, frases, "1", "1.0"), ponemos un 1
    return 1

# Aplicamos la función a todas las columnas de extras
for col in extras:
    df_red[col] = df_red[col].apply(limpiar_extras_letras)
```

```python

# 6. ELIMINAR CUALQUIER NULO RESTANTE Y EXPORTAR
df_red = df_red.dropna()
df_red.to_csv('DATASET_RED_NEURONAL_LISTO.csv', index=False)

print("--- ¡DATASET LISTO PARA LA IA! ---")
print(f"Filas finales: {len(df_red)}")
print("Todas las columnas son ahora NUMÉRICAS:")
print(df_red.dtypes)
display(df_red.head())
```

```python
import pandas as pd
import numpy as np
import os
import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard

# Cargar dataset final
df = pd.read_csv('DATASET_RED_NEURONAL_LISTO.csv')

# Separar características (X) de la variable objetivo (y)
X = df.drop('price', axis=1)
y = df['price']

# Dividir en Train (70%), Validation (15%) y Test (15%)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, random_state=42) # 0.1765 de 85% es aprox 15%

# Escalar los datos (usamos el escalador entrenado SOLO con Train para no hacer trampas)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print(f"Datos de Entrenamiento: {X_train_scaled.shape}")
print(f"Datos de Validación: {X_val_scaled.shape}")
print(f"Datos de Test: {X_test_scaled.shape}")
```

```python
def construir_modelo(tasa_dropout):
    modelo = Sequential([
        Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        Dropout(tasa_dropout), # Capa para evitar Overfitting
        Dense(64, activation='relu'),
        Dropout(tasa_dropout),
        Dense(32, activation='relu'),
        Dense(1) # Salida lineal porque predecimos un precio continuo
    ])
    
    # Compilamos usando Error Cuadrático Medio (MSE) y monitorizando Error Absoluto Medio (MAE)
    modelo.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return modelo

# Vamos a probar tres tasas de Dropout distintas
tasas_dropout = [0.2, 0.5, 0.8]
epoca_entrenamiento = 50 

for tasa in tasas_dropout:
    print(f"\n--- Entrenando modelo con Dropout: {tasa} ---")
    modelo = construir_modelo(tasa)
    
    if tasa == 0.2: # Solo lo imprimimos una vez para verificar la restricción de parámetros
        modelo.summary() 
    
    # Configurar ModelCheckpoint para guardar solo el mejor modelo
    ruta_guardado = f"mejor_modelo_dropout_{tasa}.keras"
    checkpoint = ModelCheckpoint(
        filepath=ruta_guardado,
        monitor='val_loss', # Guardar cuando la pérdida de validación mejore
        save_best_only=True,
        verbose=0
    )
    
    # Configurar TensorBoard
    log_dir = os.path.join("logs", "fit", f"dropout_{tasa}_{datetime.datetime.now().strftime('%Y%m%d-%H%M%S')}")
    tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
    
    # Entrenar
    historia = modelo.fit(
        X_train_scaled, y_train,
        epochs=epoca_entrenamiento,
        batch_size=64,
        validation_data=(X_val_scaled, y_val),
        callbacks=[checkpoint, tensorboard_callback],
        verbose=1 # Ponlo en 0 si no quieres que te llene la pantalla de texto
    )
```

```python
from tensorflow.keras.models import load_model

# 1. Cargar el mejor modelo (elige el archivo .keras del experimento que haya ganado en TensorBoard)
# Supongamos que ganó el de 0.2
mejor_modelo = load_model('mejor_modelo_dropout_0.2.keras')

# 2. Evaluar primero con el Test Ciego normal
evaluacion = mejor_modelo.evaluate(X_test_scaled, y_test, verbose=0)
print(f"Error Medio Absoluto (MAE) en Test: {evaluacion[1]:.2f} €")

# 3. EL TEST DE LA CALLE (Tus 3 pisos de Pamplona)
# ORDEN DE LAS COLUMNAS: m2_real, room_num, bath_num, condition, floor, lift, garage, swimming_pool, air_conditioner, terrace, loc_zone_value

# IMPORTANTE: Debes buscar qué 'loc_zone_value' corresponde a Pamplona en tu diccionario original (precio_medio_por_zona)
# Para este ejemplo, pondremos un valor inventado medio (ej: 250000)
VALOR_ZONA_PAMPLONA = 250000 

pisos_pamplona = pd.DataFrame([
    # Piso 1: 90m2, 3 habs, 2 baños, buen estado(1), planta 3, ascensor(1), sin garaje(0), sin piscina(0), sin AA(0), con terraza(1)
    [90, 3, 2, 1, 3, 1, 0, 0, 0, 1, VALOR_ZONA_PAMPLONA],
    
    # Piso 2: 120m2, 4 habs, 2 baños, obra nueva(2), planta 1, ascensor(1), con garaje(1), con piscina(1), con AA(1), con terraza(1)
    [120, 4, 2, 2, 1, 1, 1, 1, 1, 1, VALOR_ZONA_PAMPLONA],
    
    # Piso 3: 60m2, 1 hab, 1 baño, a reformar(0), planta 0, sin ascensor(0), sin garaje(0), sin piscina(0), sin AA(0), sin terraza(0)
    [60, 1, 1, 0, 0, 0, 0, 0, 0, 0, VALOR_ZONA_PAMPLONA]
], columns=X.columns)

# Escalar los datos de la calle con el MISMO escalador del entrenamiento
pisos_pamplona_scaled = scaler.transform(pisos_pamplona)

# Predecir
predicciones = mejor_modelo.predict(pisos_pamplona_scaled)

print("\n--- RESULTADOS DEL TEST DE LA CALLE ---")
for i, pred in enumerate(predicciones):
    print(f"Piso {i+1} en Pamplona: Precio estimado por la IA -> {pred[0]:,.2f} €")
```
